```python
import polars as pl
from polars import selectors as cs
```

```python
df_datos = pl.read_csv("../../data/dataset.csv")
```

```python
# 1
(
    df_datos.select(cs.float())
    .with_columns(
        cs.by_name(
            "Patient Age",
            "Mother's age",
            "Father's age",
            "No. of previous abortion"
        ).cast(pl.Int32)
    )
)
```

```python
# 2
df_datos.with_columns(
    cs.string().str.replace(r"^-$", "Unknown")
)
```

```python
# 3
def get_coordinate(location, coord = "latitude"):
    if location == "-":
        return None
    
    lines = location.split("\n")
    lat_long_str = lines[2].replace("(", "").replace(")", "")
    if coord == "latitude":
        coord_value = lat_long_str.split(",")[0].strip()
    else:  # longitude
        coord_value = lat_long_str.split(",")[1].strip()

    return float(coord_value)

df_datos.with_columns(
    Address = pl.col("Location of Institute").str.split("\n").list.get(0),
    Latitude = pl.col("Location of Institute").map_elements(
        lambda loc: get_coordinate(loc, "latitude"),
        return_dtype=pl.Float64
    ),
    Longitude = pl.col("Location of Institute").map_elements(
        lambda loc: get_coordinate(loc, "longitude"),
        return_dtype=pl.Float64
    ),
)
```

```python
# 4
df_datos.with_columns(
    cs.by_name(
        "Genes in mother's side",
        "Inherited from father",
        "Maternal gene",
        "Paternal gene",
        "Parental consent",
        "Folic acid details (peri-conceptional)",
        "H/O serious maternal illness",
        "Assisted conception IVF/ART",
        "History of anomalies in previous pregnancies",
    ).fill_null("No") == "Yes"
)
```

```python
# 5
df_datos.with_columns(
    year_of_birth = (pl.lit(2025) - pl.col("Patient Age")).cast(pl.Int32)
)
```

```python
# 6
df_datos.with_columns(
    all_tests_positive = pl.all_horizontal(cs.starts_with("Test") == 1)
)
```

```python
# 7
(
    df_datos.with_columns(
        (
            pl.col("White Blood cell count (thousand per microliter)")
            >= pl.col("White Blood cell count (thousand per microliter)").quantile(0.9)
        ).alias("wbc_count_high"),
        (
            pl.col("Blood cell count (mcL)")
            >= pl.col("Blood cell count (mcL)").quantile(0.9)
        ).alias("rbc_count_high"),
    )
    .filter(pl.col("wbc_count_high") & pl.col("rbc_count_high"))
)
```
