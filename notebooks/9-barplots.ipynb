```python
df_anage_subset = df_anage.select("Sample size","Specimen origin")
df_anage_gby = df_anage_subset.group_by("Sample size","Specimen origin").len()
```

```python
df_anage_gby = df_anage_gby.pivot(on="Specimen origin", index="Sample size")
df_anage_gby_pandas = df_anage_gby.to_pandas().set_index("Sample size")
```

```python
df_anage_gby_pandas
```

```python
df_anage_gby_pandas.plot(kind="bar", stacked=True)
```

```python
df_anage_gby_pandas.plot(kind="bar", stacked=True, width=0.5)
```

```python
df_anage_gby_pandas.plot(kind="bar", stacked=True, width = 1, color = ["cadetblue","salmon"], linewidth=1)
plt.title("Distribución de tamaño de muestreo por origen de los organismos")
plt.xlabel("Tamaño de muestreo")
plt.ylabel("count")
```

```python
df_anage_class = df_anage.select("Class").group_by("Class").len()
df_anage_class
```

```python
plt.pie(df_anage_class["len"], labels=df_anage_class["Class"])
```

```python
df_anage_class_filter = df_anage_class.filter(pl.col("len")>10)
```

```python
plt.pie(df_anage_class_filter["len"], labels= df_anage_class_filter["Class"])
```

```python
plt.pie(df_anage_class_filter["len"], labels= df_anage_class_filter["Class"], colors=["cadetblue","steelblue","darkseagreen","salmon","lightpink","aliceblue","plum","skyblue"])
```

```python
plt.pie(df_anage_class_filter["len"], labels= df_anage_class_filter["Class"], colors=["cadetblue","steelblue","darkseagreen","salmon","lightpink","aliceblue","plum","skyblue"], autopct="%d")
```

```python
plt.pie(df_anage_class_filter["len"], labels= df_anage_class_filter["Class"], colors=["cadetblue","steelblue","darkseagreen","salmon","lightpink","aliceblue","plum","skyblue"], autopct="%f")
```

```python
plt.pie(df_anage_class_filter["len"], labels= df_anage_class_filter["Class"], colors=["cadetblue","steelblue","darkseagreen","salmon","lightpink","aliceblue","plum","skyblue"], autopct="%.2f")
```

```python
def autopct(pct):
    if(pct>5): 
        return '%.2f' % pct
    else:
        return ""

plt.pie(df_anage_class_filter["len"], labels= df_anage_class_filter["Class"], colors=["cadetblue","steelblue","darkseagreen","salmon","lightpink","aliceblue","plum","skyblue"], autopct=autopct)

plt.title("Cantidad de organismos por clase", loc="left", fontweight="bold")
```
