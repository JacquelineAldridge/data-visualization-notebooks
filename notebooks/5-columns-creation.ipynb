```python
def get_latitude(location):
    if location == "-":
        return None
    location_list = location.split("\n")
    location_str = location_list[2]
    location_str = location_str.replace("(","")
    latitude_str = location_str.split(",")[0]
    
    return float(latitude_str)

df_datos_lat = df_datos.with_columns(
    (pl.col("Location of Institute").map_elements(get_latitude, return_dtype=pl.Float32).alias("latitude")))
```
